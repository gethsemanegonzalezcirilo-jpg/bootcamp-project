```python
import pandas as pd
df = pd.read_csv('/content/sample_data/mxmh_survey_results.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create scatter plot for Hours per day vs. Anxiety
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Anxiety', y='Hours per day', alpha=0.5)
plt.title('# of Hours of music listened per Day vs. Anxiety')
plt.xlabel('Anxiety')
plt.ylabel('Hours per Day')
plt.grid(True)
plt.show()

# Create scatter plot for Hours per day vs. Depression
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Depression', y='Hours per day', alpha=0.5)
plt.title('# of Hours of music Listene per Day vs. Depression')
plt.xlabel('Depression')
plt.ylabel('Hours per Day')
plt.grid(True)
plt.show()

# Create scatter plot for Hours per day vs. Insomnia
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Insomnia', y='Hours per day', alpha=0.5)
plt.title('# of Hours of music listened per Day vs. Insomnia')
plt.xlabel('Insomnia')
plt.ylabel('Hours per Day')
plt.grid(True)
plt.show()
```
